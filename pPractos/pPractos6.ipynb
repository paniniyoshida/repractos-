Создание линейных графиков

Построения графика, отображаем зависимость между переменными в датасете. График помогает визуализировать и анализировать данные, чтобы понять, есть ли какая-либо связь между этими переменными.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Чтение данных из файла
data = pd.read_csv('file.csv')

# Замена переменных x и y значениями из датасета
x = data['math score']
y = data['reading score']

# Построение графика
plt.plot(x, y, label='График')
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.grid()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Чтение данных из файла
data = pd.read_csv('file.csv')

# Извлечение значений из датасета
x = data['math score']
y = data['reading score']

# Построение графика
plt.plot(x, y, color='cyan', marker='o', markersize=7, label='График')
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.grid()
plt.legend()
plt.show()
```

Создание столбчатой диаграммы

Создания столбчатых диаграмм, отображаем значения переменных по оси x и по оси y. Эта диаграмма позволяет визуально сравнить значения двух переменных и сделать выводы о возможной связи между ними или о распределении данных. 

```python
import matplotlib.pyplot as plt

data = pd.read_csv('file.csv')

# Извлечение значений из датасета
x = data['reading score']
y = data['math score']


plt.bar(x,y, label = 'Продажи', color = 'purple')
plt.xlabel('Месяц')
plt.ylabel('Прибыль')
plt.title('Диаграмма')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

data = pd.read_csv('file.csv')

x = data['reading score']
y = data['math score']

plt.bar(x,y, label = 'Продажи', color = 'purple')
plt.plot(x,y, label = 'Стабильнось')
plt.xlabel('Месяц')
plt.ylabel('Прибыль')
plt.title('Диаграмма')
plt.legend()
plt.show()
```

Создание круговой диаграммы

Создание круговой диаграммы. Диаграммы отображают популярность автомобильных марок. Каждая марка представлена своим процентным соотношением от общего количества.

```python
vals = [24, 53, 21, 21, 35]
labels = ["Exeed", "Chery", "Changan", "Geely", "Jetour"]

plt.pie(vals, labels=labels, autopct='%1.1f%%>' )

plt.title("Популяность автомобильных марок")
plt.show( )
```

```python
vals = [24, 53, 44, 34, 15]
labels = ["BMW", "Toyota", "Audi", "Lexus", "Kia"]

plt.pie(vals, labels=labels, autopct='%1.1f%%>' )

plt.title("Самые аварийные автомобильные марки")
plt.show( )
```

Создание гистограммы   

Создание гистограммы, отображаем значения переменных по оси x. Визуализация гистограммы позволяет наглядно представить распределение данных и оценить их концентрацию в различных диапазонах значений.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data = pd.read_csv('file.csv')


x = data['math score'] 


plt.hist(x, bins=10, color='green', edgecolor='black')
plt.title('Гистограмма')

plt.xlabel('Значения')
plt.ylabel('Частота')

plt.show()
```
